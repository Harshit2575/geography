```python
import pandas as pd
import PyPDF2
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import os
```

```python
lis = os.listdir("/home/harshit/geography/reports2018incog/")
districts = pd.DataFrame(pd.read_csv("/home/harshit/geography/districts.csv"))
states = districts["State"]
states = set(states)
for i in lis:
    pdfFileObj = open("/home/harshit/geography/reports2018incog/"+i,'rb')    #Opening annual report pdf
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)                             #reading pdf   
    num_pages = pdfReader.numPages                                           # counting no. of pages  
    count = 0                                                            
    text = ""
    while count < num_pages:                                                #extracting text from pdf  
        pageObj = pdfReader.getPage(count)
        count +=1
        text += pageObj.extractText()
    tokens = word_tokenize(text)
    punctuations = ['(',')',';',':','[',']',',']
    stop_words = stopwords.words('english')
    keywords = [word for word in tokens if not word in stop_words and not word in punctuations]
    score_district = pd.DataFrame(columns=['District','count'])
    for p in districts["District"]:
        count = 0
        for j in keywords:
            j=j.upper()
            if p==j: count+=1
        df = pd.DataFrame({"District":[p],"count":[count]}) 
        score_district=score_district.append(df,ignore_index=True)
    score_district.to_csv("/home/harshit/geography/score_district/"+i+".csv",index=False)
     
    score_state = pd.DataFrame(columns=['State','count'])
    for s in states:
        count=0
        for q in keywords:
            q=q.upper()
            if s==q: count+=1
        df = pd.DataFrame({"State":[s],"count":[count]}) 
        score_state=score_state.append(df,ignore_index=True)
    score_state.to_csv("/home/harshit/geography/score_state/"+i+".csv",index=False)
```
