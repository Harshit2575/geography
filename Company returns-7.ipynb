```python
import os
import pandas as pd
import statsmodels.api as sm
import pickle
```

```python
#regressing company returns with hqstate mean returns
factors = pd.read_csv("/home/harshit/geography/fourfactors.csv")
lis = os.listdir("/home/harshit/geography/hqportfolios/")
mret_path = "/home/harshit/geography/hqportfolios/"
for i in lis:
    data = pd.read_csv("/home/harshit/geography/hqportfolios/"+i)
    hqret = pd.read_csv("/home/harshit/geography/meanret/"+i)
    for j in data["Company"]:
        j=j.upper()
        comp_ret = pd.read_csv("/home/harshit/geography/mret/"+j+".csv")
        if(len(comp_ret)==62):
            merged = pd.merge(factors,comp_ret,left_index=True,right_index=True)
            merged = pd.merge(merged,hqret[["meanHQ"]],left_index=True,right_index=True)
            merged["Return"] = merged["Return"]-merged["Rf"]
            X= merged[["SMB","HML","WML","Rm-Rf","meanHQ"]]
            X = sm.add_constant(X)
            y = merged["Return"]
            
            model1= sm.OLS(y,X,hasconst=True).fit()
            model1_name = "/home/harshit/geography/hqcompmodels/withmeanhq/"+j+".sav"
            pickle.dump(model1,open(model1_name,"wb"))
            X= merged[["SMB","HML","WML","Rm-Rf"]]
            X = sm.add_constant(X)
            y = merged["Return"]
            model2= sm.OLS(y,X,hasconst=True).fit()
            model2_name = "/home/harshit/geography/hqcompmodels/withoutmeanhq/"+j+".sav" 
            pickle.dump(model2,open(model2_name,"wb"))
```

```python
# Getting Summary of company return prediction models
for i in os.listdir("/home/harshit/geography/hqcompmodels/withoutmeanhq/"):
    model1 = pickle.load(open("/home/harshit/geography/hqcompmodels/withoutmeanhq/"+i,"rb"))
    model2 = pickle.load(open("/home/harshit/geography/hqcompmodels/withmeanhq/"+i,"rb"))
    print(i.split(".")[0])
    print("\nSummary for prediction of meanER with meanHQ\n")
    print(model1.summary())
    print("Summary for prediction of meanER with meanHQ\n")
    print(model2.summary())
```
